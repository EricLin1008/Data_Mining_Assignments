#### Exercise 4: Dataframes
In the following, we will analyze UFO sightings over the last century.

````python
import pandas as pd

df = pd.read_csv('data/ufo.csv')
df.head()
````

##### __(a) Print the total number of UFO sightings.__

````python
df.shape[0]
````

##### __(b) Print all sightings of UFOs in Berlin, Germany where the shape is not unknown.__

````python
x = df[(df['country'] == 'de') & (df['city'] == 'berlin (germany)') & (df['shape'] != 'unknown')]
x
````

##### __(c) Print the total number of sightings per country and the total number of sightings for each shape.__

````python
country = df['country'].value_counts()
shape = df['shape'].value_counts()

print("Total Number of Sightings per Country:")
print(country)
print("\nTotal Number of Sightings per Shape:")
print(shape)
````

##### __(d) By using the results from (c) check how often the most sighted shape was seen in the country with the most sightings.__

````python
max_country = country.idxmax()
max_shape = shape.idxmax()

max_country_with_max_shape = df[(
    df['country'] == max_country) & (df['shape'] == max_shape)].shape[0]

print(f"Country with the most sightings: {max_country}")
print(f"Most sighted shape: {max_shape}")
print(f"How often the most sighted shape was seen in the country with the most sightings: {max_country_with_max_shape} times")
````

##### __(e) Print the mean and the median duration of UFO sightings in seconds.__ _(Hint: Find a reasonable way to deal with non-numeric values in the data)_

````python
df.info()
````

````python
df['duration (seconds)'] = pd.to_numeric(df['duration (seconds)'], errors='coerce')

mean = df['duration (seconds)'].mean()
median = df['duration (seconds)'].median()

print(f"Mean Duration (seconds): {mean:.3f}")
print(f"Median Duration (seconds): {median:.3f}")
````

##### __Why is there such a big difference between the two values?__

```
>> In this dataset, the mean significantly exceeding the median indicates the presence of extreme values, or outliers.
```
